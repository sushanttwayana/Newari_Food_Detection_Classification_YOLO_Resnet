```python
from ultralytics import YOLO


# Load the trained model
model = YOLO("/kaggle/input/yolo-with-resnet-model/best.pt")
model
```

```python
# Run detection on a new image
results = model.predict(source="/kaggle/input/yolo-with-resnet-model/alu.png", save=True, imgsz=512, conf=0.5)

# Show the results
results[0].show()
```
